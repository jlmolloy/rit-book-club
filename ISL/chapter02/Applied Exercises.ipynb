# Question 8
This exercise relates to the `College` data set.

* `Private` : Public/private indicator
* `Apps` : Number of applications received
* `Accept` : Number of applicants accepted
* `Enroll` : Number of new students enrolled
* `Top10perc` : New students from top 10 % of high school class
* `Top25perc` : New students from top 25 % of high school class
* `F.Undergrad` : Number of full-time undergraduates
* `P.Undergrad` : Number of part-time undergraduates
* `Outstate` : Out-of-state tuition
* `Room`.Board : Room and board costs
* `Books` : Estimated book costs
* `Personal` : Estimated personal spending
* `PhD` : Percent of faculty with Ph.D.’s
* `Terminal` : Percent of faculty with terminal degree
* `S`.F.Ratio : Student/faculty ratio
* `perc`.alumni : Percent of alumni who donate
* `Expend` : Instructional expenditure per student
* `Grad.Rate` : Graduation rate

### 8 (c) iv.
Create a new qualitative variable, called `Elite`, by binning
the `Top10perc` variable. We are going to divide universities
into two groups based on whether or not the proportion
of students coming from the top 10% of their high school
classes exceeds 50%.

```r
# Create an array of boolean values
Elite <- rep(FALSE, nrow(college))
Elite[college$Top10perc > 50] <- TRUE
# Append to data frame
college <- data.frame(college, Elite)
```

```r
summary(college["Elite"])
```

```r
boxplot(Outstate~Elite, data=college)
```

### 8 (c) v.
Use the `hist()` function to produce some histograms with
differing numbers of bins for a few of the quantitative variables.

https://www.rdocumentation.org/packages/graphics/versions/3.6.2/topics/hist

You may find the command `par(mfrow = c(2, 2))`
useful: it will divide the print window into four regions so
that four plots can be made simultaneously. Modifying the
arguments to this function will divide the screen in other
ways.

```r
hist(college$Personal,
     breaks=6,
     main="Estimated personal spending", xlab='Spending/USD', ylab='Students')
```

```r
par(mfrow = c(2, 2))
hist(college$Enroll,
     breaks=10,
     main="Number of new students enrolled", xlab='Students')
hist(college$Terminal,
     breaks=10,
     main="Faculty with terminal degree")
hist(college$Expend,
     breaks=10,
     main="Instructional expenditure per studen")
hist(college$Grad.Rate,
     breaks=10,
     main="Graduation Rate")
```

### 8 (d) vi.
Continue exploring the data, and provide a brief summary
of what you discover

```r
# Log-log scatter plot of expenditure versus gradutation rate
data <- log(college[c("Grad.Rate", "Expend")])
plot(Grad.Rate~Expend, data=data, main="Expediture versus graduation rate")
abline(lm(Grad.Rate~Expend, data=data))
```
